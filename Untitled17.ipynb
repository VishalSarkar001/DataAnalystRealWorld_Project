```python
def split_clean_reject(df, rules):

    if not rules:
        return df.copy(), pd.DataFrame(columns=list(df.columns) + ["reject_reason"])

    reject_df = pd.DataFrame(index=df.index)

    for rule_name, condition in rules:
        condition = condition.reindex(df.index, fill_value=False)
        reject_df[rule_name] = np.where(condition, "", rule_name)

    reject_reason_col = reject_df.apply(
        lambda x: ",".join(x[x != ""]),
        axis=1
    )

    rejected_df = df.loc[reject_reason_col != ""].copy()
    rejected_df["reject_reason"] = reject_reason_col[reject_reason_col != ""]

    clean_df = df.loc[reject_reason_col == ""].copy()

    return clean_df, rejected_df
```

```python
# Load clean datasets for analytics (DO NOT overwrite raw data)
clean_customers_df = pd.read_csv("clean_customers.csv")
clean_sales_df = pd.read_csv("clean_sales_header.csv")
clean_loyalty_rules_df = pd.read_csv("clean_loyalty_rules.csv")

print(clean_customers_df.head())
print(clean_sales_df.head())
print(clean_loyalty_rules_df.head())
```

```python
import numpy as np

# Select default loyalty rule (lowest spend threshold)
default_rule = clean_loyalty_rules.loc[
    clean_loyalty_rules["min_spend_threshold"].idxmin()
]

points_rate = default_rule["points_per_unit_spend"]
min_spend = default_rule["min_spend_threshold"]
bonus = default_rule["bonus_points"]

# Calculate earned loyalty points per transaction
sales_with_loyalty_points = clean_sales_header.copy()

sales_with_loyalty_points["earned_points"] = np.where(
    sales_with_loyalty_points["total_amount"] >= min_spend,
    (sales_with_loyalty_points["total_amount"] * points_rate) + bonus,
    0
)

sales_with_loyalty_points[
    ["transaction_id", "customer_id", "total_amount", "earned_points"]
].head()
```

```python
# Aggregate loyalty points at customer level
customer_loyalty_points = (
    sales_with_loyalty_points
    .groupby("customer_id", as_index=False)["earned_points"]
    .sum()
)

customer_loyalty_points.head()
```

```python
# Merge loyalty points into customer master
customers_with_loyalty = clean_customers.merge(
    customer_loyalty_points,
    on="customer_id",
    how="left"
)

# Customers with no transactions get 0 points
customers_with_loyalty["earned_points"] = (
    customers_with_loyalty["earned_points"].fillna(0)
)

customers_with_loyalty.head()
```

```python
# Assign loyalty status based on earned points
customers_with_loyalty["loyalty_status"] = pd.cut(
    customers_with_loyalty["earned_points"],
    bins=[-1, 999, 4999, float("inf")],
    labels=["Bronze", "Silver", "Gold"]
)

customers_with_loyalty.head(50)
```

```python
# Ensure transaction_date is datetime
clean_sales_header["transaction_date"] = pd.to_datetime(
    clean_sales_header["transaction_date"],
    errors="coerce"
)

# Reference date (latest transaction date)
analysis_date = clean_sales_header["transaction_date"].max()

# Build RFM metrics
rfm = (
    clean_sales_header
    .groupby("customer_id")
    .agg(
        recency=("transaction_date", lambda x: (analysis_date - x.max()).days),
        frequency=("transaction_id", "count"),
        monetary=("total_amount", "sum")
    )
    .reset_index()
)

rfm.head()
```

```python
# Merge RFM metrics into final customer table
final_customer_table = customers_with_loyalty.merge(
    rfm,
    on="customer_id",
    how="left"
)

final_customer_table.head(50)
```

```python
import numpy as np

final_customer_table["customer_segment"] = np.select(
    [
        (final_customer_table["loyalty_status"] == "Gold") |
        (
            (final_customer_table["monetary"] >= 5000) &
            (final_customer_table["frequency"] >= 3)
        ),

        final_customer_table["loyalty_status"] == "Silver",

        (final_customer_table["recency"] > 90) &
        (final_customer_table["frequency"] <= 2)
    ],
    [
        "High Value",
        "Loyal",
        "At Risk"
    ],
    default="New / Low Value"
)

final_customer_table.head(50)
```

```python
final_customer_table = final_customer_table.drop(
    columns=[
        "total_loyalty_points",
        "last_purchase_date",
        "segment_id"
    ],
    errors="ignore"
)

final_customer_table.head(50)
```
