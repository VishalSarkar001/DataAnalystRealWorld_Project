```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
customers = pd.read_csv('/content/customer_details_with_unclean_data.csv')
stores = pd.read_csv('/content/stores_with_unclean_data.csv')
products = pd.read_csv('/content/products_with_unclean_data.csv')
loyalty_rules = pd.read_csv('/content/loyalty_rules_with_unclean_data.csv')
sales_header = pd.read_csv('/content/store_sales_header_with_unclean_data.csv')
sales_line_items = pd.read_csv('/content/store_sales_line_items_with_unclean_data.csv')

print("All datasets loaded successfully")
```

```python
datasets = {
    "stores": stores,
    "products": products,
    "customers": customers,
    "loyalty_rules": loyalty_rules,
    "sales_header": sales_header,
    "sales_line_items": sales_line_items
}

for name, df in datasets.items():
    print(f"\n{name.upper()}")
    print("Rows:", df.shape[0])
    print("Columns:", df.shape[1])
    print(df.columns.tolist())
```

```python
for name, df in datasets.items():
    print(f"\n{name.upper()}")
    print("Total NULL values:", df.isna().sum().sum())
    print("Duplicate rows:", df.duplicated().sum())
```

```python
def split_clean_reject(df, rules):
    """
    rules: list of tuples -> (rule_name, boolean_condition)
    """
    reject_reasons = []

    for rule_name, condition in rules:
        failed = ~condition
        reject_reasons.append(
            failed.map(lambda x: rule_name if x else "")
        )

    reject_reason_col = (
        pd.concat(reject_reasons, axis=1)
        .apply(lambda x: ",".join(filter(None, x)), axis=1)
    )

    rejected_df = df[reject_reason_col != ""].copy()
    rejected_df["reject_reason"] = reject_reason_col[reject_reason_col != ""]

    clean_df = df[reject_reason_col == ""].copy()

    return clean_df, rejected_df
```

```python
store_rules = [
    ("NULL_STORE_ID", stores["store_id"].notna()),
    ("NULL_STORE_NAME", stores["store_name"].notna()),
    ("INVALID_OPENING_DATE", pd.to_datetime(stores["opening_date"], errors="coerce") <= pd.Timestamp.today()),
    ("NULL_CITY", stores["store_city"].notna()),
]

clean_stores, rejected_stores = split_clean_reject(stores, store_rules)
```

```python
product_rules = [
    ("NULL_PRODUCT_ID", products["product_id"].notna()),
    ("NULL_PRODUCT_NAME", products["product_name"].notna()),
    ("NEGATIVE_PRICE", products["unit_price"] > 0),
    ("NEGATIVE_STOCK", products["current_stock_level"] >= 0),
]

clean_products, rejected_products = split_clean_reject(products, product_rules)
```

```python
customer_rules = [
    ("NULL_CUSTOMER_ID", customers["customer_id"].notna()),
    ("INVALID_EMAIL", customers["email"].astype(str).str.contains("@")),
    ("INVALID_PHONE", customers["customer_phone"].astype(str).str.len() == 10),
    ("NEGATIVE_LOYALTY_POINTS", customers["total_loyalty_points"] >= 0),
    ("FUTURE_LAST_PURCHASE", pd.to_datetime(customers["last_purchase_date"], errors="coerce") <= pd.Timestamp.today()),
]

clean_customers, rejected_customers = split_clean_reject(customers, customer_rules)
```

```python
loyalty_rules_checks = [
    ("NULL_RULE_ID", loyalty_rules["rule_id"].notna()),
    ("NEGATIVE_POINTS_RATE", loyalty_rules["points_per_unit_spend"] >= 0),
    ("NULL_MIN_SPEND", loyalty_rules["min_spend_threshold"].notna()),
]

clean_loyalty_rules, rejected_loyalty_rules = split_clean_reject(loyalty_rules, loyalty_rules_checks)
```

```python
sales_header_rules = [
    ("NULL_TRANSACTION_ID", sales_header["transaction_id"].notna()),
    ("NEGATIVE_TOTAL_AMOUNT", sales_header["total_amount"] > 0),
    ("NULL_TRANSACTION_DATE", sales_header["transaction_date"].notna()),
    ("INVALID_CUSTOMER_ID", sales_header["customer_id"].isin(clean_customers["customer_id"])),
    ("INVALID_STORE_ID", sales_header["store_id"].isin(clean_stores["store_id"])),
]

clean_sales_header, rejected_sales_header = split_clean_reject(sales_header, sales_header_rules)
```

```python
line_item_rules = [
    ("INVALID_QUANTITY", sales_line_items["quantity"] > 0),
    ("NEGATIVE_LINE_AMOUNT", sales_line_items["line_item_amount"] > 0),
    ("INVALID_PRODUCT_ID", sales_line_items["product_id"].isin(clean_products["product_id"])),
]

clean_sales_line_items, rejected_sales_line_items = split_clean_reject(sales_line_items, line_item_rules)
```

```python
sales["transaction_date"] = pd.to_datetime(
    sales["transaction_date"],
    errors="coerce"
)

sales = sales.dropna(subset=["transaction_date"])
```

```python
analysis_date = sales["transaction_date"].max()
analysis_date
```

```python
rfm = sales.groupby("customer_id").agg(
    recency=("transaction_date", lambda x: (analysis_date - x.max()).days),
    frequency=("transaction_id", "count"),
    monetary=("total_amount", "sum")
).reset_index()

rfm.head()
```

```python
rfm = rfm.dropna()
```

```python
rfm["R_rank"] = rfm["recency"].rank(method="first", ascending=False)
rfm["F_rank"] = rfm["frequency"].rank(method="first", ascending=True)
rfm["M_rank"] = rfm["monetary"].rank(method="first", ascending=True)

rfm["R_score"] = pd.qcut(rfm["R_rank"], 4, labels=[4,3,2,1])
rfm["F_score"] = pd.qcut(rfm["F_rank"], 4, labels=[1,2,3,4])
rfm["M_score"] = pd.qcut(rfm["M_rank"], 4, labels=[1,2,3,4])

rfm["RFM_score"] = (
    rfm["R_score"].astype(int) +
    rfm["F_score"].astype(int) +
    rfm["M_score"].astype(int)
)

rfm.head()
```
